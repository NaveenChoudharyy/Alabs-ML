## Key Definitions:
Invoice Value: Invoice Value =(( Quantity*Avg_price)*(1-Dicount_pct)*(1+GST))+Delivery_Charges

Average order value = Revenue / Transaction per customer

Profit margin is the commonly used profitability ratio. It represents how much percentage of total sales has earned as the gain.

Purchase Frequency is the average number of purchases made by a customer over a defined period of time (typically one month or one year). It is the sum of total number transactions divided by total number customers.

Repeat rate shows you the percentage of your current customer base that has come back to shop again.

Churn Rate is the annual percentage rate at which customers stop subscribing.

Customer lifetime value, lifetime customer value, or life-time value is a prediction of the net profit/revenue attributed to the entire future relationship with a customer.

## Other inputs related to Analysis for additional reference:

### 1. Why do we need customer Segmentation?
As every customer is unique and can be targeted in different ways. The Customer segmentation plays an important role in this case. The segmentation helps to understand profiles of customers and 
can be helpful in defining cross sell/upsell/activation/acquisition strategies.

### 2. What is RFM Segmentation?
RFM Segmentation is an acronym of recency, frequency and monetary based segmentation. 

Recency is about when the last order of a customer. It means the number of days since a customer made the last purchase. If it’s a case for a website or an app, this could be interpreted as the last 
visit day or the last login time.

Frequency is about the number of purchases in a given period. It could be 3 months, 6 months or 1 year. So we can understand this value as for how often or how many customers used the product of 
a company. The bigger the value is, the more engaged the customers are. 

Alternatively

We can define, average duration between two transactions

Monetary is the total amount of money a customer spent in that given period. Therefore big spenders will be differentiated with other customers such as MVP or VIP.

### 3. What is LTV and How to define it?
In the current world, almost every retailer promotes its subscription and this is further used to understand the customer lifetime. Retailer can manage these customers in better manner if they know which customer is high life time value.

Customer lifetime value (LTV) can also be defined as the monetary value of a customer relationship, based on the present value of the projected future cash flows from the customer relationship. Customer lifetime value is an important concept in that it encourages firms to shift their focus from quarterly profits to the long-term health of their customer relationships. Customer lifetime value is an important metric because it represents an upper limit on spending to acquire new customers. For this reason it is an important element in calculating payback of advertising spent in marketing mix modelling.

#### 4. Why do need to predict Customer Lifetime Value?

The LTV is an important building block in campaign design and marketing mix management. Although targeting models can help to identify the right customers to be targeted, LTV analysis can help to quantify the expected outcome of targeting in terms of revenues and profits. The LTV is also important because other major metrics and decision thresholds can be derived from it. For example, the LTV is naturally an upper limit on the spending to acquire a customer, and the sum of the LTVs for all of the customers of a brand, known as the customer equity, is a major metric for business valuations. Similarly to many other problems of marketing analytics and algorithmic marketing, LTV modelling can be approached from descriptive, predictive, and prescriptive perspectives.

### 5. How Next Purchase Day helps to Retailers?
Our objective is to analyse when our customer will purchase products in the future so for such customers we can build strategy and can come up with strategies and marketing campaigns accordingly.

a. Group-1: Customers who will purchase in more than 60 days

b. Group-2: Customers who will purchase in 30-60 days

c. Group-3: Customers who will purchase in 0-30 days 

### 6. What is Cohort Analysis? How it will be helpful?
A cohort is a group of users who share a common characteristic that is identified in this report by an Analytics dimension.

For example, all users with the same Acquisition Date belong to the same cohort. 

The Cohort Analysis report lets you isolate and analyze cohort behaviour.

Cohort analysis in e-commerce means to monitor your customers’ behaviour based on common traits they share – the first product they bought, when they became customers, etc. - - to find patterns and tailor marketing activities for the group

```python
# UDF for analysis - for continuous variable only
def continuous_var_summary( x ):
    
    # freq and missings
    n_total = x.shape[0]
    n_miss = x.isna().sum()
    perc_miss = n_miss * 100 / n_total
    
    # outliers - iqr
    q1 = x.quantile(0.25)
    q3 = x.quantile(0.75)
    iqr = q3 - q1
    lc_iqr = q1 - 1.5 * iqr
    uc_iqr = q3 + 1.5 * iqr
    
    return pd.Series( [ x.dtype, x.nunique(), n_total, x.count(), n_miss, perc_miss,
                       x.sum(), x.mean(), x.std(), x.var(), 
                       lc_iqr, uc_iqr, 
                       x.min(), x.quantile(0.01), x.quantile(0.05), x.quantile(0.10), 
                       x.quantile(0.25), x.quantile(0.5), x.quantile(0.75), 
                       x.quantile(0.90), x.quantile(0.95), x.quantile(0.99), x.max() ], 
                     
                    index = ['dtype', 'cardinality', 'n_tot', 'n', 'nmiss', 'perc_miss',
                             'sum', 'mean', 'std', 'var',
                        'lc_iqr', 'uc_iqr',
                        'min', 'p1', 'p5', 'p10', 'p25', 'p50', 'p75', 'p90', 'p95', 'p99', 'max']) 





```

```python
# As per my understanding, the problem is that we have to find the retained customer from the last month
# as retention is asked on month on month basis



# If a customer is purchasing in consucutive month then only we can say him/her as retained customer
# If a customer is purchasing but having a gap of more then one month then he/she is not retained on monthly basis
```

```python
# Creating 2 new columns from column "transaction_Date" as "month" and "month_name"

df_22 = df[['customerID', 'month', 'month_name']]
```

### Calculate the Revenue, Marketing spend, percentage of marketing spend out of revenue, Tax, percentage of delivery charges by month.


```python
# Creating a dataset with required variables

df_27 = df[['month', 'month_name', 'invoice_value','delivery_Charges', 'gst', 'marketing_spend', 'online_Spend', 'offline_Spend']]
```

```python
#

df_27 = df_27.groupby(['month', 'month_name', 'marketing_spend', 'gst','online_Spend', 'offline_Spend'])[['invoice_value', 'delivery_Charges']].sum().reset_index()
```

```python
# Calculating gst amount

df_27['gst_amount'] = df_27.invoice_value*(df_27.gst+1)*df_27.gst
```

```python
df_27 = df_27.groupby(['month','month_name'])[['marketing_spend', 'online_Spend', 'offline_Spend', 'invoice_value', 'delivery_Charges', 'gst_amount']].sum()

df_27 = df_27.reset_index().drop(columns='month')
```

```python
# Creating new variables


df_27['percentage_of_marketing_spend_out_of_revenue'] = df_27.marketing_spend/df_27.invoice_value *100
df_27['percentage_of_delivery_charges_by_month'] = df_27.delivery_Charges/df_27.invoice_value *100
df_27['percentage_of_gst_amont_out_of_revenue'] = df_27.gst_amount/df_27.invoice_value *100
```

```python
df_27.columns = ['Month', 'Tot_maketing_spend', 'online_marketing_spend', 'offline_marketing_spend', 'Revenue', 'Tot_delivery_charges', 'Tot_gst_amount', 'percentage_of_marketing_spend_out_of_revenue','percentage_of_delivery_charges_by_month','percentage_of_gst_amont_out_of_revenue']
```

```python
# Visualization

fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(15, 30))
fig.suptitle("Trends/Seasonality of sales by location", fontsize=20)

axes = axes.ravel()

for i, column in enumerate(df_27.columns[1:6]):
    ax = axes[i]
    ax.bar(df_27['Month'], df_27[column], color=sns.color_palette('gist_earth'))
    ax.set_title(column)
    ax.set_xlabel("Month")
    ax.set_ylabel(column)
    ax.legend([column], loc='upper right')
    ax.grid()

# Remove any empty subplots
for i in range(len(df_27.columns[1:6]), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.subplots_adjust(top=1)
plt.show()
```

# 3. Performing Customer Segmentation

### Heuristic (Value based, RFM) – Divide the customers into Premium, Gold, Silver, Standard customers and define strategy on the same. 

### Value based segmentation
ideally, for value based segmentation we should divide the data based on location for better results but here all the cities are from same country so i am not doing this to save my time.

```python
# Today's date 

today = 5112023
today = pd.to_datetime(today, format='%d%m%Y')
```

```python
# To get receny we need to find the last date of transaction for each customer

df_31 = df[['customerID', 'date']].sort_values(by='date', 
                                               ascending=True).drop_duplicates(subset='customerID', keep = 'last')
```

```python
# Finding the recency by subracting the given date with todays date and getting no. of days in return

df_31['recency'] = (today - df_31.date).dt.days
df_31.head()
```

```python
# For frequency and monetory we need to use groupby on the customer

df_31a = df.groupby('customerID').agg({'invoice_value':'sum', 'transaction_ID': 'count'}).reset_index()

df_31a.columns = ['customerID', 'monetary', 'frequency']
df_31a.head()
```

```python
# Now mergeing the two dataset to get a new dataset

df_31b = pd.merge(left=df_31, right=df_31a, left_on=['customerID'], right_on=['customerID'], how='inner')
df_31b.head()
```

```python
# Creating deciles

df_31b['decile'] = pd.qcut(df_31b.monetary, 10, labels=False)
df_31b.head()
```

```python
# Using groupby function to remove duplicacy in decile variable

df_31b = df_31b.groupby('decile')[[ 'monetary','recency', 'frequency']].mean().reset_index()
df_31b


print('Now the data is divided into 10 segmentation based on the monetary, this is called value based segmentation')
```

### RFM segmentation
ideally, for value based segmentation we should divide the data based on location for better results but here all the cities are from same country so i am not doing this to save my time.

```python
#Creating a new dataset containing 'customerID' and 'date' and removing duplicates

df_31c = df[['customerID','date']].sort_values(by='date', ascending=True).drop_duplicates(subset='customerID', keep='last')
df_31c.head()
```

```python
# Creaing another dataset containing total revenue and total transactions per customer

df_31d = df.groupby('customerID').agg({'invoice_value': 'sum', 'transaction_ID':'count'}).reset_index()
df_31d.head()
```

```python
# merging the two datasets for further analysis

df_31e = pd.merge(left=df_31c, right=df_31d, left_on='customerID', right_on='customerID', how='inner')
df_31e.head()
```

```python
#Creaing a new variable named as recency contining number of days from today
df_31e['recency'] = (today - df_31e.date).dt.days

# dropping variable 'date' as not required
df_31e = df_31e.drop(columns='date')

# correcting the column names
df_31e.columns = ['customerID', 'monetary', 'frequency', 'recency']
df_31e.head()
```

```python
# Creating 3 new variables from existing variables

df_31e['r_seg']= pd.qcut(df_31e.recency, 4, labels=False)
df_31e['f_seg']= pd.qcut(df_31e.frequency, 4, labels=False)
df_31e['m_seg']= pd.qcut(df_31e.monetary, 4, labels=False)
df_31e.head()
```

```python
# Encoding the recency variable as per the customer last purchase date as 'active' or lesser active 

df_31e.loc[df_31e.r_seg == 0, 'r_seg'] = 'active'
df_31e.loc[df_31e.r_seg == 1, 'r_seg'] = 'less active'
df_31e.loc[df_31e.r_seg == 2, 'r_seg'] = 'at risk'
df_31e.loc[df_31e.r_seg == 3, 'r_seg'] = 'churn'
```

```python
# Defining the rules for segmenting the customers

df_31e['fm_seg'] = np.where(((df_31e.f_seg == 3) & (df_31e.m_seg == 3)), 'Premium', 
        np.where(((df_31e.f_seg == 2) & (df_31e.m_seg == 3)) | ((df_31e.f_seg == 3) & (df_31e.m_seg == 2)), 'Gold',
        np.where(((df_31e.f_seg == 1) & (df_31e.m_seg == 3))| ((df_31e.f_seg == 3) & (df_31e.m_seg == 1))|
                ((df_31e.f_seg == 2) & (df_31e.m_seg == 2)), 'Silver', 'Standard')))
df_31e.head()
```

```python
df_31e.fm_seg.value_counts()
```

```python
# Using groupby to further divide the data on the bais of customer recency

df_31e.groupby(['r_seg', 'fm_seg'])[['frequency','recency','monetary']].mean()
```

```python
# Required package

from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df.isna().sum().sum()

# no null value in any column
```

```python
# cReating a pivot table to know the products sold per transaction

cs_tab = df.pivot_table(index='transaction_ID', columns='product_SKU', values = 'quantity', 
                        aggfunc='count').fillna(0)
cs_tab.head()
```

```python
# Crating a function 

def encode(x):
    if x <= 0:
        return 0
    else:
        return 1
```

```python
# Appling the function

cs_tab = cs_tab.applymap(encode)
```

```python
# Correcting columns names

cs_tab.columns = cs_tab.columns.str.replace(' ', '_').str.replace('.', '_')
cs_tab.columns = cs_tab.columns.str.replace('-', '_').str.replace('+', '_')
cs_tab.columns = cs_tab.columns.str.replace('/', '_')
cs_tab.columns
```

# 6. Predicting Next Purchase Day(How soon each customer can visit the store (0-30 days, 30-60 days, 60-90 days, 90+ days)
#

###### For this, we need create dependent variable at customer level (average days per one transaction for only repeat customers and divide into groups 0-30 days, 30-60 days, 60-90 days and 90+ days) then build classification model to predict next purchase of given customer.


```python
## For this, I am assuming that if a customer has made more then 1 transaction in a single day
## then it will be considerd as non-repeat customer, no matter if the transaction_ID is same or different
## Only transactions made on different days will be considerd as repeat customer




#####---- Not including non-repeat customers for this analysis as 
```

```python
# Creating the new dataset with the required variables

data_a = df[['customerID', 'date','gender', 'location', 'tenure_Months', 'product_Category', 'quantity',
       'unit_price', 'delivery_Charges', 'coupon_Status', 'gst','discount_pct', 'offline_Spend', 
             'online_Spend', 'month', 'invoice_value']]
```

```python
data_b = data_a[['customerID', 'date']].drop_duplicates().sort_values(by=['customerID','date'])


# dropping subset = ['customer', 'date'] as next purchase days is asked and many 
##  customers has made multiple transactions in a single day
```

```python
# Using shift formula to get the transaction date and customer name from the next row.


data_b['next_date'] = data_b.date.shift(-1)
data_b['next_customer'] = data_b.customerID.shift(-1)
```

```python
# If a customerID and next_customer are same it means we can take the difference of the 
## date and  next_date to get number of days between next transaction of same customer
### If a customer has made a single transaction it will be 0 and also from the last 
#### transaction of repeat customer it will also be 0

data_b = data_b.dropna().loc[data_b.customerID == data_b.next_customer, :]
data_b['gap_days'] = (data_b.next_date - data_b.date).dt.days
```

```python
# Dropping the unnecessary variables 

data_b = data_b.drop(columns=['next_date', 'next_customer'])
data_b
```

```python
# Merging the two datasets to get a new dataset


data_c = pd.merge(left = data_a, right= data_b, on= ['customerID', 
                                                     'date'], how='left').drop(columns=
                                                                               ['customerID', 'date']).fillna(91)
data_c.head()

# As 'customerID' has very high cardinality and cannot be used for modelling
## For this analysis those who made sinle transactions i am assuming that those 
###will make transaction minimum after 90 days so filling 91 days in nulls.
```

```python
## Now "gap_days" is y_variable for this classification problem so we need to encode this


data_c['gap_days'] = np.where(data_c.gap_days <= 30, 0, 
                              np.where(data_c.gap_days <= 60, 1, 
                                       np.where(data_c.gap_days <= 90, 2, 3)))

data_c.head(1)
```

```python
# Converting column month, gst and discount_pct as 'object' for the analysis as having order and low cardinality

data_c.month = data_c.month.astype('object')
data_c.gst = data_c.gst.astype('object')
data_c.discount_pct = data_c.discount_pct.astype('object')
```

# Perform cohort analysis by defining below cohorts

#### Customers who started in each month and understand their behaviour
#### Which Month cohort has maximum retention?


```python
# Creating a new dataset with only required variables and sorting as per customerID and then as per month

tab = df[['customerID', 'month']].drop_duplicates()



# Creating a new variable having first month of transaction of respective customer

tab = tab.sort_values(['customerID', "month"])


# Creating a new variable as customer tenure by subracting 
## the fist transaction month by respective transaction month of customer 

tab['aq_month'] = tab.groupby('customerID')['month'].transform('min')



# Creating a new veriable having minimum tenure for each respective customer

tab['trans_month'] = tab.month - tab.month.min()


# Creating a new variable cohort_month by getting minimum trans_month for each respactive customer

tab['cohort_month'] = tab.groupby('customerID')['trans_month'].transform('min')
```

```python
# Creating a new variable cohort_index 

tab['cohort_index'] = tab.trans_month - tab.cohort_month



# Getting count of customers for both 'cohort_index' and 'cohort_month' 
tab = tab.groupby(['cohort_month', 'cohort_index'])[['customerID']].count().reset_index()

# Correcting column names
tab.columns = ['cohort_month','cohort_index','number_of_customers']
```

```python
tab = tab.pivot_table(columns='cohort_index', values='number_of_customers', index='cohort_month', aggfunc='sum')
tab
```

```python
# Converting table into percentage

tab = tab.divide(tab.iloc[:, 0], axis=0).round(3)*100
```

```python
# Visualization

fig, ax = plt.subplots(1,1, figsize=(16,10))
sns.heatmap(tab, annot=True, cmap='BuPu', fmt='g')
ax.xaxis.set_label_position('top')
plt.show()
```
